```python
df = df.dropna().reset_index(drop=True)
```

```python
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
y = df["price_range"]
```

```python
x = df.drop("price_range",axis=1)
```

```python
x_train,x_val,y_train,y_val = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
result = {}
for i in range(1001):
    regressor = RandomForestClassifier(n_estimators=i+1,random_state=i,oob_score=True) 
    regressor.fit(x_train, y_train)
    ypred = regressor.predict(x_val)
    print(f"#Est: {i+1} randStat: {i} ========== ",r2_score(y_val,ypred))
    result[r2_score(y_val,ypred)] = f"#Est: {i+1} randStat: {i}"
```

```python
key = max(list(result.keys()))
val = result[key]
print(val)
```

```python
regressor = RandomForestClassifier(n_estimators=368,random_state=367,oob_score=True) 
regressor.fit(x_train, y_train)
```
